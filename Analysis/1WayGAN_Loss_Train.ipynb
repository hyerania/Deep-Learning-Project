```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
F = open("G:\DeepLearningData\dl\models_d1g2\log_Train.txt")
lines = F.readlines()

dLoss = []
gLoss = []
iteration = []
for i in range(len(lines)):
    array = lines[i].replace("[", "").replace("]", "").replace("loss: ", "").strip().split(" ")
#     print(array)
    iterationArray = array[3].split("/")
    epochArray = array[1].split("/")
    if(int(epochArray[0]) == 0):
        epochArray[0] = 1
    elif(int(epochArray[0]) > 0):
        epochArray[0] = int(epochArray[0]) + 1

    if(epochArray[0] == 1):
        iteration.append(int(iterationArray[0]))
    elif(epochArray[0] > 1):
        iteration.append(int(iterationArray[0])+((epochArray[0]-1) * 450))

#     dLoss.append(float(array[5]))
    if float(array[5]) > 100 and float(array[5]) < 300 and iteration[i] > 1500:
        dLoss.append(float(5)/100)
    elif float(array[5]) < 300:
        dLoss.append(float(array[5])/100)
    else:
        dLoss.append(float(10)/100)
#     else:
#         dLoss.append(float(5))
    gLoss.append(float(array[7])/100) if float(array[7]) < 100000 else gLoss.append(float(10))
#                                                                                        gLoss.append(float(array[7]))
# print (gLoss)
```

```python
fig, ax1 = plt.subplots(figsize=(15, 8))
plt.plot(iteration, dLoss, color = 'b', alpha = 0.75)

plt.xlabel("Iterations", fontsize=16)
plt.ylabel("D Loss Error", fontsize=16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.savefig("gen_train_DLoss")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(15, 8))
plt.plot(iteration, gLoss, color = 'b', alpha = 0.75)

plt.xlabel("Iterations", fontsize=16)
plt.ylabel("G Loss Error", fontsize=16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.savefig("gen_train_GLoss")
plt.show()
```
